```python
import random

import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import truncnorm
import numpy as np
```

```python
fig, ax = plt.subplots(figsize=(10, 5))
sns.distplot([random.gauss(0, 1) for _ in range(1000000)], ax=ax)
ax.set_xlabel("X")
ax.set_ylabel("Densidad")
plt.savefig("aux_normal.png")
plt.show()
```

# **Truncnorm**

```python
loc, scale = 200, 25
a, b = -1, 1

samples = truncnorm.rvs(a, b, loc=loc, scale=scale, size=10000, random_state=0).round().astype(int)
sns.displot(samples);
```

# **Activation functions**

```python
def tanh(x):
    return np.tanh(x)

def relu(x):
    return np.maximum(0, x)

def sigmoid(x):
    return 1 / (1 + np.exp(-x))
```

```python
plt.rcParams.update({'font.size': 14})
plt.rcParams['lines.linewidth'] = 3

fig, ax = plt.subplots(figsize=(20, 5), ncols=3)

x = np.linspace(-10, 10, 100)
ax[0].plot(x, tanh(x))
ax[0].set_title("tanh(x)")
ax[0].set_xticks([-10, -5, 0, 5, 10])
ax[0].set_xticklabels(["$-\infty$", "$-5$", "$0$", "$5$", "$\infty$"])

ax[1].plot(x, relu(x))
ax[1].set_title("ReLU(x)")
ax[1].set_xticks([-10, -5, 0, 5, 10])
ax[1].set_xticklabels(["$-\infty$", "$-5$", "$0$", "$5$", "$\infty$"])

ax[2].plot(x, sigmoid(x))
ax[2].set_title("$\sigma$(x)")
ax[2].set_xticks([-10, -5, 0, 5, 10])
ax[2].set_xticklabels(["$-\infty$", "$-5$", "$0$", "$5$", "$\infty$"])

plt.savefig("aux_activations.png", bbox_inches="tight")
plt.show()
```
