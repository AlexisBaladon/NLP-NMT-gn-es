# **Data augmentation**

```python
import os

import pandas as pd

from utils import plots, constants
```

```python
DATASET_FOLDER = constants.AUGMENTATION_FOLDER
TUNING_FOLDER = constants.TUNING_FOLDER

DATASET_PATH_DEFAULT_S2S_GN_ES_AUGMENTED = os.path.join(DATASET_FOLDER, 'default_s2s_gn_es.csv')
DATASET_PATH_DEFAULT_S2S_ES_GN_AUGMENTED = os.path.join(DATASET_FOLDER, 'default_s2s_es_gn.csv')
DATASET_PATH_DEFAULT_TRANSFORMER_GN_ES_AUGMENTED = os.path.join(DATASET_FOLDER, 'default_transformer_gn_es.csv')
DATASET_PATH_DEFAULT_TRANSFORMER_ES_GN_AUGMENTED = os.path.join(DATASET_FOLDER, 'default_transformer_es_gn.csv')
LVL1_RESULTS = os.path.join(TUNING_FOLDER, 'lvl1_augmented.csv')

FIGSIZE = (20, 6)
```

```python
df_default_s2s_gn_es_augmented = pd.read_csv(DATASET_PATH_DEFAULT_S2S_GN_ES_AUGMENTED)
df_default_s2s_es_gn_augmented = pd.read_csv(DATASET_PATH_DEFAULT_S2S_ES_GN_AUGMENTED)
df_default_transformer_gn_es_augmented = pd.read_csv(DATASET_PATH_DEFAULT_TRANSFORMER_GN_ES_AUGMENTED)
df_default_transformer_es_gn_augmented = pd.read_csv(DATASET_PATH_DEFAULT_TRANSFORMER_ES_GN_AUGMENTED)
df_default_transformer_es_gn_augmented = df_default_transformer_es_gn_augmented[~df_default_transformer_es_gn_augmented['model_name'].str.endswith('vocab.npz')]

lvl1_df = pd.read_csv(LVL1_RESULTS)
df_default_gn_es = lvl1_df[lvl1_df['source'].str.contains('gn', regex=False)]
df_default_es_gn = lvl1_df[lvl1_df['source'].str.contains('es', regex=False)]

df_default_s2s_gn_es = df_default_gn_es[df_default_gn_es['model_name'].str.contains('s2s')]
df_default_s2s_es_gn = df_default_es_gn[df_default_es_gn['model_name'].str.contains('s2s')]
df_default_transformer_gn_es = df_default_gn_es[df_default_gn_es['model_name'].str.contains('transformer')]
df_default_transformer_es_gn = df_default_es_gn[df_default_es_gn['model_name'].str.contains('transformer')]
```
