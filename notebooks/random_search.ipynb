```python
import random; SEED=42
import collections

from scipy.stats import loguniform, uniform, truncnorm, randint
import matplotlib.pyplot as plt
import seaborn as sns
```

- **discrete**: uniform (random.choice)
- **continuous**: truncnormal/loguniform/uniform

## **Loguniform**

```python
a, b = 0.0001, 1.0
samples = list(loguniform.rvs(a, b, size=10000, random_state=SEED))
sns.displot(data=samples);
print(samples[:10])
```

## **Normal** (truncated)

```python
loc, scale = 200, 25
a, b = -3, 3

samples = truncnorm.rvs(a, b, loc=loc, scale=scale, size=1000, random_state=SEED).round().astype(int)
sns.displot(data=samples);
print(samples[:10])
```

```python
loc, scale = 7, 1.5
a, b = -3, 3

samples = truncnorm.rvs(a, b, loc=loc, scale=scale, size=10000, random_state=SEED).round().astype(int)
sns.displot(data=samples);

samples_counter = collections.Counter(samples)
max_depth = 6

samples_counts_bellow_max_depth = {sample: sample_count for sample, sample_count in samples_counter.items() if sample <= max_depth}
samples_counter = {sample: round(sample_count / sum(samples_counts_bellow_max_depth.values()), 4) for sample, sample_count in samples_counts_bellow_max_depth.items()}
sorted(list(samples_counter.items()), key=lambda x:x[0])
```

## **Multinoulli**

```python
population_domain = ["2000 2000", "4000 4000", "6000 6000", "8000 8000", "10000 10000", "12000 12000"]
weights = [0.05, 0.1, 0.3, 0.3, 0.1, 0.05]
ramdom_instance = random.Random(); ramdom_instance.seed(SEED)

samples = ramdom_instance.choices(population_domain, weights, k=100000)
samples = list(sorted(map(lambda s: int(s.split()[0]), samples)))
sns.displot(samples);
```

## **Uniform**

```python
a, b = 0.01, 1.05
samples = list(uniform.rvs(a, b, size=100000, random_state=SEED))
sns.displot(data=samples);
```

```python
a, b = 1, 10
samples = list(randint.rvs(a, b, size=100000, random_state=SEED))
sns.displot(data=samples);
```
