```python
import os
import sys
import warnings; warnings.filterwarnings('ignore')

import pandas as pd

utils_path = os.path.join(os.getcwd(), '..', '..', '..')
if utils_path not in sys.path:
    sys.path.insert(0, utils_path)

from notebooks.src.utils import plots, constants
```

```python
AUGMENTATION_FOLDER = constants.AUGMENTATION_FOLDER
PRETRAINING_FOLDER = constants.PRETRAINING_FOLDER
TUNING_FOLDER = constants.TUNING_FOLDER

TUNING_FOLDER_CLEANED = os.path.join(TUNING_FOLDER, 'cleaned')
TUNING_FOLDER_RAW = os.path.join(TUNING_FOLDER, 'raw')

AUGMENTATION_DEFAULT_CSV = os.path.join(AUGMENTATION_FOLDER, 'default.csv')
FINETUNED_RESULTS_CSV = os.path.join(TUNING_FOLDER_CLEANED, 'best_models.csv')
DEFAULT_RESULTS_CSV = os.path.join(PRETRAINING_FOLDER, 'default_pretraining.csv')
PRETRAINED_RESULTS_CSV = os.path.join(PRETRAINING_FOLDER, 'finetuned_pretraining.csv')

FIGSIZE = (20, 4)
```

```python
augmentation_default_df = pd.read_csv(AUGMENTATION_DEFAULT_CSV)

original_default_df = pd.read_csv(DEFAULT_RESULTS_CSV)
original_default_df['pretraining_epochs'] = original_default_df['model_name'].str.extract(r'^.*finetuning-epochs_([0-9]+)\.npz')
original_default_df = original_default_df[original_default_df['pretraining_epochs']=='0']

default_df = pd.concat([augmentation_default_df, original_default_df], axis=0) \
               .drop('pretraining_epochs', axis=1)

default_df.head()
```

```python
default_s2s_gn_es_df = default_df[default_df['model_name'].str.contains('s2s') & default_df['source'].str.contains('gn')]
default_s2s_es_gn_df = default_df[default_df['model_name'].str.contains('s2s') & default_df['source'].str.contains('es')]
default_transformer_gn_es_df = default_df[default_df['model_name'].str.contains('transformer') & default_df['source'].str.contains('gn')]
default_transformer_es_gn_df = default_df[default_df['model_name'].str.contains('transformer') & default_df['source'].str.contains('es')]
```
